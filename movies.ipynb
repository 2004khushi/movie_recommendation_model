## for tagline -> Facts:

```python
df['overview'] = df['overview'].fillna('')
df['tagline'] = df['tagline'].fillna('')
df = df.dropna(subset=['title'])

df[['overview','tagline','title']].isnull().sum()
```

```python
df= df.drop_duplicates().reset_index(drop=True)
df.duplicated().sum()
```

```python
# now let's see what's in genre and if it needs some cleaning
df['genres'][0]
```

```python
#to hume isme ye faaltu ki id thodi chhaiye we just need the genre in it!
import ast
df['genres'] = df['genres'].apply(lambda x :" ".join([i['name'] for i in ast.literal_eval(x)]))
df['genres'].head()
```

```python
df['genres'].isnull().sum()
```

```python
df['tags'] = df['overview'] + df['tagline'] + df['genres']
df['tags'][0]
```
