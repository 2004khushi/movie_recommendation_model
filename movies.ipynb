```python
#data cleaning for nlp ->

# 1) removal of punctuations
# 2) removal of stop words
# 3) lemmatizer (ending and end is same so ending ko end(uske root word) pe lejaayenge!)

import nltk
from nltk.corpus import stopwords
import re
from nltk.stem import WordNetLemmatizer
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess_text(text):
  #1) lower everyone taaki when and When ko alag na dekhajaaye
  text = str(text).lower()

  #2) remove punctuattions
  text = re.sub(r'[^a-zA-Z\s]', '',text)

  #3) word tokenization
  words = text.split()

  #4) remove stop words
  words = [word for word in words if word not in stop_words]

  #5) Lemmatize
  words = [lemmatizer.lemmatize(word) for word in words]

  #6) join words
  text = ' '.join(words)

  return text
```

```python
df['tags'] = df['tags'].apply(preprocess_text)
```

```python
df['tags'][0]
```

```python
# reset_index
df = df.reset_index(drop=True)
indices = pd.Series(df.index, index=df['title'])
indices
```

```python
# vectorization
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf  = TfidfVectorizer(max_features=50000,ngram_range=(1,2),stop_words='english')
tfidf_matrix = tfidf.fit_transform(df['tags'])
tfidf_matrix.shape
```

```python
print(tfidf.get_feature_names_out())
```

## Cosine Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def recommend(title, n=10):
    if title not in indices:
        return ['Movie not found']

    idx = indices[title]

    sim_score = cosine_similarity(
        tfidf_matrix[idx],
        tfidf_matrix
    ).flatten()

    sim_idx = sim_score.argsort()[::-1][1:n+1]

    return df['title'].iloc[sim_idx]
```

```python
recommend('Toy Story')
```

## Applying filters and ranks

**STEP 1: Base similarity** <br />
This gives you candidates, not final recommendations.

```python
def get_similar_indices(title, top_k=50):
    idx = indices[title]

    sim_scores = cosine_similarity(
        tfidf_matrix[idx],
        tfidf_matrix
    ).flatten()

    # exclude the movie itself
    similar_indices = sim_scores.argsort()[::-1][1:top_k+1]

    return similar_indices, sim_scores
```

**STEP 2: SECONDARY / CONDITIONAL LOGIC** <br />

(Franchise boost, availability, safety) <br />

These are hard rules.

```python
def apply_secondary_filters(candidate_indices):
    filtered = []

    for i in candidate_indices:
        # must be released
        if df.loc[i, 'status'] != 'Released':
            continue

        # remove adult content
        if df.loc[i, 'adult'] == True:
            continue

        filtered.append(i)

    return filtered
```

**STEP 3: OPTIONAL USER FILTERS** <br />

(language, runtime) <br />

These depend on user preference, not similarity.

```python
def apply_user_filters(
    candidate_indices,
    allowed_languages=None,
    max_runtime=None
):
    filtered = []

    for i in candidate_indices:
        if allowed_languages is not None:
            if df.loc[i, 'original_language'] not in allowed_languages:
                continue

        if max_runtime is not None:
            if pd.isna(df.loc[i, 'runtime']) or df.loc[i, 'runtime'] > max_runtime:
                continue

        filtered.append(i)

    return filtered
```

**STEP 4: RANKING (this is the most important part)**

We now order, not remove.

Simple & correct ranking formula:

*   similarity â†’ primary
*   vote_average â†’ quality
*   popularity â†’ discoverability
*   vote_count â†’ confidence

```python
df['vote_average'] = pd.to_numeric(df['vote_average'], errors='coerce')
df['vote_count'] = pd.to_numeric(df['vote_count'], errors='coerce')
df['popularity'] = pd.to_numeric(df['popularity'], errors='coerce')

def rank_candidates(candidate_indices, sim_scores):
    ranked = []

    for i in candidate_indices:
        similarity = sim_scores[i]

        vote_avg = df.loc[i, 'vote_average']
        vote_cnt = df.loc[i, 'vote_count']
        popularity = df.loc[i, 'popularity']

        # handle NaNs safely
        vote_avg = 0 if pd.isna(vote_avg) else vote_avg
        vote_cnt = 0 if pd.isna(vote_cnt) else vote_cnt
        popularity = 0 if pd.isna(popularity) else popularity

        score = (
            0.6 * similarity +
            0.2 * vote_avg +
            0.15 * np.log1p(vote_cnt) +
            0.05 * popularity
        )

        ranked.append((i, score))

    ranked.sort(key=lambda x: x[1], reverse=True)
    return [i for i, _ in ranked]
```

**STEP 5: FINAL RECOMMEND FUNCTION**

```python
def recommend_with_filters(
    title,
    n=10,
    allowed_languages=None,
    max_runtime=None
):
    if title not in indices.index:
        return ['Movie not found']

    # 1. similarity
    candidate_indices, sim_scores = get_similar_indices(title)

    # 2. secondary filters
    candidate_indices = apply_secondary_filters(candidate_indices)

    # 3. user filters
    candidate_indices = apply_user_filters(
        candidate_indices,
        allowed_languages=allowed_languages,
        max_runtime=max_runtime
    )

    # 4. ranking
    ranked_indices = rank_candidates(candidate_indices, sim_scores)

    return df['title'].iloc[ranked_indices[:n]].values
```

```python
recommend_with_filters(
    'Toy Story',
    allowed_languages=['en'],
    max_runtime=120
)
```

```python
recommend_with_filters('Superstar Goofy')
```
